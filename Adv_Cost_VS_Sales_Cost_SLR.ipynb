Importing the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Accessing the dataset

```python
dataset = pd.read_csv('Advertising_Sales_SLR.csv')
x = dataset.iloc[:,:-1].values
#print(x)
y = dataset.iloc[:,-1].values
#print (y)
```

Splitting the data into training set and test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

Train the SLR model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

Visualizing the train set

```python
plt.scatter(x_train, y_train, color='red') #the dots
plt.plot(x_train, regressor.predict(x_train), color='blue') #the line
plt.title('Sales vs Advertising cost')
plt.xlabel('Advertising cost')
plt.ylabel('Sales')
plt.show()
```

Visualizing the test set results

```python
plt.scatter(x_test, y_test, color = 'green',label='Actual Test Data')
plt.plot(x_train, regressor.predict(x_train), color = 'blue', label='Regression Line')
# NOTE: Line is always plotted from training data.
plt.title('Sales vs Advertising cost')
plt.show()
```

Model performance

```python
from sklearn.metrics import r2_score, mean_squared_error

y_pred = regressor.predict(x_test)

print("R²:", r2_score(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("RMSE:", mean_squared_error(y_test, y_pred) ** 0.5)
#
#R² (Coefficient of Determination) → How much X --> Y variation is explained by the model (0–1, higher = better)
#MSE (Mean Squared Error) → “Model's average squared error” (unit = y², lower = better)
#RMSE (Root Mean Squared Error) → “Model's average error in original units of y” (√MSE, lower = better, easy to interpret)
```
